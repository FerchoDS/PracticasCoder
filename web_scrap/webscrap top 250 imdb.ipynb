```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.imdb.com/chart/top/')
#soup = BeautifulSoup(r.text, 'lxml')
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
tag_object=soup.title
tag_object
```

```python
eq = soup.find_all('td',class_='titleColumn') 
print(eq)
```

```python
titulos = list ()
nombre = list()
for i in eq:
    titulos.append(i.text)
   
print(titulos)
```

```python
for body in eq:
  if body.find('div', {'class': 'date'}):# grab div by class name
    text = body.find('td',class_='titleColumn')
    # find div by 'title' attribute
    title2 = text.get('title','No title attribute')
    print(title2)
```

```python
import pandas as pd

df = pd.DataFrame(titulos)
```

```python
#Preparacion de los datos

df = df[0].str.split('[\n]', expand=True) #separo texto en columnas con separados /n
df = df.drop([0, 4], axis=1) #elimino columnas 0 y 4
df.set_axis(['pos', 'titulo','year'], axis='columns', inplace=True) #asigno nombre a las columnas
df['titulo'] = df['titulo'].str.strip() #saco los espacios adelante y detras
df['year'] = df['year'].str.strip('()') #saco los parentesis de la columna año
df['pos'] = df['pos'].str.strip('.') # saco el punto de la Pos
df
```

```python
df.to_excel("250mejoresIMDB.xlsx", index=False) #genero un excel con las 250 posiciones
```
